### Defining Symbols

```python
from sympy import symbols, Function
x, y, z, t = symbols('x y z t')
k, m, n = symbols('k m n', integer=True)
f, g, h = symbols('f g h', cls=Function)
```

```python
# Finding limit of function
from sympy import sin, cos, limit
f = sin(x)/x
limit(f,x,0)
```

```python
from sympy import diff
f = sin(x)/x
g = diff(f, x)
g
```

```python
% matplotlib inline

from sympy.plotting import plot

f = sin(x)/x
g = diff(f, x)

p1 = plot(f, show=False)
p2 = plot(g, line_color='red', show=False)
p1.append(p2[0])
p1.show()
```

### Differentiating using Definition

```python
x, h = symbols('x h')
f = x**n
Δf = (x+h)**n - x**n
limit(Δf/h, h, 0)
```

```python
f = lambda x: sin(x) * cos(x)
def Δ(f):
    return f(x+h) - f(x)
limit(Δ(f)/h, h, 0)
```

```python
p3 = plot(sin(h)/h, (h, -20, 20), show=False)
p3.show()
```

```python
def f(x):
    return sin(x) * cos(x)
limit((f(x+h) - f(x))/h, h, 0)
```

```python
from sympy import diff
diff(sin(x) * cos(x), x)
```

## Integration

```python
from sympy import integrate
f = sin(x) * cos(x)
integrate(f, x)
```
